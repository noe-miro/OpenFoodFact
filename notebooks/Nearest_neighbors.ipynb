```python
import pickle
import numpy as np

#dimension reduction
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP

# clustering 
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity

#affichage
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as mpimg
import pandas as pd
```

```python
with open('../data/embeddings/emb_light.pckl', 'rb') as f:
    emb_light = pickle.load(f)
```

```python
x_umap = np.array(list(emb_light.values()))
print(x_umap.shape)
filenames = list(emb_light.keys())
print(len(filenames))
```

```python
# Calculate the cosine similarity matrix
# cosine_sim = cosine_similarity(x)
cosine_sim = cosine_similarity(x_umap)
print(cosine_sim.shape)
```

```python
df_similarity = pd.DataFrame(cosine_sim, columns=filenames, index=filenames).astype('float16')
print(len(df_similarity.columns))
df_similarity.head()
```

SAVE DATAFRAME TO GET NEAREST NEIGHBORS LATER

```python
with open('../data/df_similarity.pckl', 'wb') as f:
    pickle.dump(df_similarity, f)
```
